```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("Pakistan_Poverty_Dataset_2000_2023.csv")
print(df.head())
```

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python

print(df.isnull().sum())


df['Year'] = pd.to_datetime(df['Year'], format='%Y')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Unemployment Rate (%)', marker='o', color='red')
plt.title("Unemployment Rate in Pakistan (2000–2023)")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data=df, x='Year', y='Unemployment Rate (%)', label='Unemployment Rate', color='red')
sns.lineplot(data=df, x='Year', y='GDP Growth Rate (%)', label='GDP Growth Rate', color='green')
sns.lineplot(data=df, x='Year', y='Inflation Rate (%)', label='Inflation Rate', color='blue')
plt.title("Unemployment vs GDP Growth vs Inflation")
plt.xlabel("Year")
plt.ylabel("Rate (%)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Economic Indicators")
plt.show()
```
